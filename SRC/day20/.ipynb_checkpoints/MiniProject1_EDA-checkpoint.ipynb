# Mini Project 1 – Exploratory Data Analysis (EDA)
## PHASE 1
This dataset contains customer analytics information. Each row represents one customer, including their demographics, income, shopping behavior, device preference, and purchase activity. The goal is to inspect, clean, and analyze the data to discover useful patterns.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('customer_analytics.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## PHASE 2 Data Cleaning

```python
df.isnull().sum()
```

Education and AnnualIncome contain missing values. I will fill:
- Education with the most frequent value (mode)
- AnnualIncome with the median (robust to outliers).

```python
df['Education'] = df['Education'].fillna(df['Education'].mode()[0])
df['AnnualIncome'] = df['AnnualIncome'].fillna(df['AnnualIncome'].median())

df = df.drop_duplicates()

df.isnull().sum()
```

## PHASE 3 Univariate Analysis

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution of Customers")
plt.show()
```

This chart shows how customer ages are distributed.

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Count')
plt.show()
```

This chart shows the number of customers by gender.

```python
df['SpendingScore'].hist()
plt.title('Spending Score Distribution')
plt.show()
```

This chart shows how spending scores vary among customers.

## Bivariate Analysis

```python
sns.scatterplot(x='AnnualIncome', y='SpendingScore', data=df)
plt.title('Income vs Spending Score')
plt.show()
```

This plot shows the relationship between income and spending.

```python
sns.boxplot(x='Gender', y='LastPurchaseAmount', data=df)
plt.title('Last Purchase by Gender')
plt.show()
```

This plot compares purchase amounts across genders.

## PHASE 4 Correlation Heatmap

```python
corr = df.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

## Executive Summary
1. Customers show a wide income range with moderate spending patterns.
2. Spending score is somewhat related to income and purchase frequency.
3. Online visits and purchase frequency may help predict customer value.
